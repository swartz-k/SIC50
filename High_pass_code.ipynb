```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
#8bit png dirctory
RAW_PNG_DIR = ('E:\\raw')
```

```python
#find edge png save dirctory
EDGE_PNG_SAVE_DIR = ('E:\\Highpass')
```

```python
def prepend(list, str):
    str += '{0}'
    list = [str.format(i) for i in list]
    return(list)

png_files = os.listdir(RAW_PNG_DIR)
png_prepend = RAW_PNG_DIR+"\\"
png_dir = prepend(png_files,png_prepend)

#print("png_files", png_files[0])
#print("png_dir", png_dir[0])
```

```python
hpf8_filter = [[-1,-1,-1], [-1,8,-1], [-1,-1,-1]]
file_count = 0
```

```python
for i in png_dir:
    img = mpimg.imread(i)
    n,m = img.shape
    hpf8_img = np.zeros_like(img)
        
    for row in range(3,n-2):
        for col in range(3,m-2):
            local_pixels = img[row-1:row+2, col-1:col+2]
        
            hpf8_transformed_pixels = hpf8_filter*local_pixels
            hpf8_score = hpf8_transformed_pixels.sum()/4
        
            #print("edge score",(edge_score)*2)
            #print("edge row col",edges_img[row,col] )
            hpf8_img[row,col]=hpf8_score
    plt.imsave(EDGE_PNG_SAVE_DIR + "\\hpf8_" + png_files[file_count], hpf8_img, cmap = 'gray')
    file_count = file_count+1
    print(file_count, "/", len(png_dir))
print("Done")
```
